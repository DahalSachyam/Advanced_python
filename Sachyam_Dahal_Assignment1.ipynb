## Question 6: List Sanitization and Metrics
Given:
```python
marks = [45, -3, 88, "NA", 102, None, 67]
```

Write a function that:
1. Removes invalid entries  
2. Returns:
   - Cleaned list  
   - Number of invalid entries  
   - Percentage of valid marks  

Return values using a tuple.


````python
marks = [45, -3, 88, "NA", 102, None, 67]

def clean_marks(lst):
    cleaned = []
    invalid = 0
    for x in lst:
        if isinstance(x, int) and x >= 0 and x <= 100: #isinstance inbuilt function check if the given input x is of type int and to detect of int number more and less than 0 and 100 we use simple condition statement
            cleaned.append(x)
        else:
            invalid += 1
    percentage = (len(cleaned) / len(lst)) * 100 #i guess precentage of valid marks means how much percentage of entires are valid in the list?
    return cleaned, invalid, percentage

clean_marks(marks)
````

## Question 8: Prime Number Function
Write `is_prime(n)` and display prime numbers between 1 and 50.


````python
def is_prime(n): #fucntion to check if the given argument is prime or not 
    if n < 2:
        return False # should be greater than 1
    i = 2
    while i * i <= n: #if i squared(50% of divsible number) itself is greater than prime there will be nothing divisibe making it prime
        if n % i == 0: # checking prime if divisible just return 0(false)
            return False
        i += 1 # increase for next iter
    return True

for x in range(1, 51):
    if is_prime(x):
        print(x)
````

## Question 9: Function Composition
Write functions to:
1. Calculate total marks  
2. Calculate average  
3. Classify result (Distinction / Pass / Fail)

Then write a function that combines them and returns:
```
(name, total, average, result)
```

Explain why modular design is beneficial.


````python
def total_marks(marks):
    s = 0
    for x in marks:
        s += x
    return s

def average_marks(total, count):
    return total / count

def classify(avg):
    if avg >= 75:
        return "Distinction"
    elif avg >= 40:
        return "Pass"
    else:
        return "Fail"

def student_report(name, marks):
    total = total_marks(marks)
    avg = average_marks(total, len(marks))
    result = classify(avg)
    return name, total, avg, result

# Example
student_report("Sita", [78, 85, 90])
#Modular design is beneficial because it breaks a complex task into small, manageable functions.
````

## Question 10: Defensive Input Parsing
Write a function that accepts a raw input string in the format:

```
"Ram|101|22|78, 88, 91"
```
Hint: Parse student data from string format: "Name|Roll|Age|Mark1, Mark2, Mark3"

Tasks:
1. Extract name, roll number, age, and marks  
2. Convert them into appropriate data types  
3. Validate age (≥16) and marks (0–100)  
4. Return the result as a dictionary using roll number as the key  

The function must not crash for malformed input.


````python
def parse_student(raw):
    try: #simple exception handling
        parts = raw.split("|") #basically on encoutering | it just splits the data in a list named parts
        if len(parts) != 4: #if its not correct format just return a empty dictionary
            return {}

        name = parts[0].strip() #extract name
        roll = parts[1].strip() #extract roll
        age = int(parts[2].strip()) #extract age as integer

        if age < 16: #check if greater than 16 or not , if not just retrun a empty dictionary
            return {}

        marks_raw = parts[3].split(",") #extract marks in a list
        marks = []
        for m in marks_raw:
            m = m.strip() #removes spaces
            if m.isdigit(): #checks if its a digit
                m = int(m) 
                if 0 <= m <= 100:
                    marks.append(m)
                else:
                    return {}
            else:
                return {}

        return {roll: {"name": name, "age": age, "marks": marks}}
    except:
        return {}

# Example
parse_student("Ram | 101 | 22 | 78, 88, 91")
````

## Question 11: Type Stability and Conversion
Given the list:
```python
data = [10, "20", 30.5, "abc", None, True]
```

Write a function that:
1. Separates numeric and non-numeric values  
2. Converts numeric values to float  
3. Returns both lists  

Explain in comments why `True` behaves as a numeric value.


````python
data = [10, "20", 30.5, "abc", None, True]

def separate_and_convert(lst):
    numeric = []
    non_numeric = []
    for x in lst:
        if isinstance(x, (int, float)) and not isinstance(x, bool): #checking if its float and not bool
            numeric.append(float(x))
        elif isinstance(x, str) and x.isdigit(): #check if its str but also a digit 
            numeric.append(float(x))
        else:
            non_numeric.append(x) #append rest
    return numeric, non_numeric

numeric_values, non_numeric_values = separate_and_convert(data)
print(numeric_values, non_numeric_values)
# True behaves as a numeric value because in Python:
# True is treated as 1 and False is treated as 0.
# This happens because bool is a subclass of int.
# Therefore, isinstance(True, int) == True.
# We exclude bool explicitly because it should not be treated as a number here.
````

## Question 13: Set-Based Enrollment Analysis
Given:
```python
python = {"Amit", "Sita", "Hari", "Nita"}
sql = {"Hari", "Ram", "Nita"}
statistics = {"Amit", "Hari", "Gita"}
```

Find:
1. Students enrolled in exactly two subjects  
2. Students enrolled in only one subject  
3. Students enrolled in all subjects


````python
python = {"Amit", "Sita", "Hari", "Nita"}
sql = {"Hari", "Ram", "Nita"}
statistics = {"Amit", "Hari", "Gita"}

all_students = python | sql | statistics #union operator so it lists all students without duplicates

exact_two = set()
only_one = set()

for s in all_students:
    count = 0
    if s in python:
        count += 1
    if s in sql:
        count += 1
    if s in statistics:
        count += 1
    if count == 2:
        exact_two.add(s)
    if count == 1:
        only_one.add(s)

all_three = python & sql & statistics #and operator finds common between all

print(exact_two, only_one, all_three)
````

## Question 14: Dictionary Aggregation Logic
Given:
```python
records = {
    "Amit": [78, 82, 91],
    "Sita": [88, 79, 85],
    "Hari": [45, 52, 60]
}
```

Write a function that:
1. Computes average per student  
2. Identifies top and bottom performers  
3. Returns results using a single dictionary


````python
records = {
    "Amit": [78, 82, 91],
    "Sita": [88, 79, 85],
    "Hari": [45, 52, 60]
}

def analyze(records):
    averages = {}
    for name in records:
        total = 0
        marks = records[name] #take marks of each std
        for m in marks: 
            total += m #find total for each student
        avg = total / len(marks) #find avg for each student
        averages[name] = avg #define dictonary with std name and averages

    top_student = None
    top_avg = -1
    bottom_student = None
    bottom_avg = 999 #intitalizations

    for name in averages:
        if averages[name] > top_avg: #comparing stds avgs to find the top_avg
            top_avg = averages[name]
            top_student = name
        if averages[name] < bottom_avg: #comaprting stds avgs to find the bottom_avg
            bottom_avg = averages[name]
            bottom_student = name

    return {
        "averages": averages,
        "top_performer": (top_student, top_avg),
        "bottom_performer": (bottom_student, bottom_avg)
    }

analyze(records)
````

## Question 15: Nested Data Structure Analysis
Given a nested dictionary representing a school's grade book:
```python
school_data = {
    "Class A": {
        "Amit": [78, 82, 91],
        "Sita": [88, 79, 85]
    },
    "Class B": {
        "Hari": [45, 52, 60],
        "Ram": [92, 88, 95]
    }
}
```

Write a function that:
1. Calculates the average marks for each student
2. Finds the top performer in each class
3. Calculates the overall school average
4. Returns a dictionary with:
   - `class_toppers`: {class_name: (student_name, average)}
   - `school_average`: float
   - `student_averages`: {student_name: average}

Explain in comments why nested dictionaries are useful for hierarchical data.

````python
school_data = {
    "Class A": {
        "Amit": [78, 82, 91],
        "Sita": [88, 79, 85]
    },
    "Class B": {
        "Hari": [45, 52, 60],
        "Ram": [92, 88, 95]
    }
}

def analyze_school(data):
    student_averages = {}                 # store average of each student
    class_toppers = {}                    # store top performer for each class
    total_sum = 0                         # total of all marks in school
    total_count = 0                        # count of all marks

    for class_name in data:                  # loop through each class
        top_student = None
        top_avg = -1                          # start lowest avg to compare

        for student in data[class_name]:     # loop through each student in class
            marks = data[class_name][student]
            s = 0
            for m in marks:
                s += m                       # sum all marks for this student

            avg = s / len(marks)        # compute student average
            student_averages[student] = avg

            if avg > top_avg:                # check if this student is class topper
                top_avg = avg
                top_student = student

            total_sum += s              # add students marks to school total
            total_count += len(marks)     # add number of marks

        class_toppers[class_name] = (top_student, top_avg)

    school_average = total_sum / total_count  # compute overall school average

    return {
        "class_toppers": class_toppers,
        "school_average": school_average,
        "student_averages": student_averages
    }

analyze_school(school_data)
# Nested dictionaries are useful for hierarchical data because:
# - They organize information in layers (school → class → student).
# - They allow easy access to data inside each group.
# - They keep related values together without long or confusing variable names.


````
